## Dealing with Null Values

```python
def fillNaMean(cols):
    for i in cols:
        data[i] = data[i].fillna(data[i].mean())

columns = ['word_count', 'char_count', 'has_images', 'has_videos','readability_score','num_shares','num_comments']
fillNaMean(columns)
```

## Dropping Irrelevant Columns

```python
data.drop('id', axis=1, inplace=True)
data.drop('state', axis=1, inplace=True)
data.drop('author', axis=1, inplace=True)
data.drop('title', axis=1, inplace=True)
data.drop('date_published', axis=1, inplace=True)
```

```python
data.info()
```

```python
def changeFloattoInt64(cols):
    for i in cols:
        data[i] = data[i].astype('int64')

columns = ['sentiment_score', 'word_count', 'char_count', 'has_images', 'has_videos', 'readability_score', 'num_shares', 'num_comments', 'clickbait_score', 'plagiarism_score']
changeFloattoInt64(columns)
```

```python
data.info()
```

```python
data['label'].unique()
```

```python
data_clean = data.copy()
```

```python
def encodeCols(cols):
    for i in cols:
        temp = pd.DataFrame({i:data[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(temp))
        data_clean[i] = data_LE.transform(data[i])

columns = ['label', 'fact_check_rating', 'political_bias','category','source','text']
encodeCols(columns)
```

```python
data_clean.to_csv(r'cleaned-data.csv', index=False, header=True)
```

```python
data_clean.head(2)
```

## Train-Test Split

```python
X = data_clean.drop('label', axis=1)  

y = data_clean['label']  

trainX, testX, trainY, testY = train_test_split(X, y, test_size=0.2, shuffle=False, random_state=42)
```

```python
trainX.head(2)
```

```python
trainY.head(2)
```

```python
print(f"Train features shape: {trainX.shape}")
print(f"Test features shape: {testX.shape}")
print(f"Train target shape: {trainY.shape}")
print(f"Test target shape: {testY.shape}")
```

## Training

```python
model_SVC = SVC()
model_SVC.fit(trainX, trainY)

print(model_SVC)
```

## Processing Results

```python
svc_predictions = model_SVC.predict(testX)
print(svc_predictions)
```

## Accuracy

```python
from sklearn.metrics import accuracy_score
svc_accuracy = accuracy_score(testY, svc_predictions)

print("-- Model Accuracy: ", end="")
print(round(svc_accuracy,3))
```
